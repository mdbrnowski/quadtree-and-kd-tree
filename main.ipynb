```python
from quadtree_visualization import QuadtreeVisualization
from geometry import Rectangle
from numpy.random import seed, uniform
```

```python
seed(0)

def generate_uniform_points(left, right, n = 10 ** 2):
    return list(zip(uniform(left, right, size=n), uniform(left, right, size=n)))
```

## Quadtree

```python
points = generate_uniform_points(-2, 2, 25)
rectangle = Rectangle(-0.3, 1, -0.7, 0.5)
```

```python
quadtree = QuadtreeVisualization(points)
quadtree.add_grid()
quadtree.find(rectangle)
quadtree.vis.show()
```

```python
quadtree.vis.show_gif(interval=1000)
```

## Kd-tree

```python
from kd_tree_visualization import KdTreeVis
```

```python
def generate_uniform_points(left, right, n = 10 ** 2):
    return list(zip(uniform(left, right, size=n), uniform(left, right, size=n)))
```

```python
def rectangle_lines_from_values(min_x, max_x, min_y, max_y):
    return Rectangle(min_x, max_x, min_y, max_y), [((min_x, min_y), (max_x, min_y)), 
                                                   ((max_x, min_y), (max_x, max_y)), 
                                                   ((max_x, max_y), (min_x, max_y)),
                                                   ((min_x, max_y), (min_x, min_y))]
```

```python
points_test = generate_uniform_points(0, 100, 100)
```

```python
kd_tree = KdTreeVis(points_test)
kd_tree.find(Rectangle(10, 80, 5, 72))
kd_tree.vis.show()
```
